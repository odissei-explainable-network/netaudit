```python
import numpy as np
import matplotlib.pyplot as plt

from scripts.edge_utilities import calc_dimension_utility
```

```python
num_dim = 32

target_dim = 2

target_dim_range = np.linspace(0, 1, 50)

source_embedding = np.array([0.5] * num_dim)
target_embedding = np.array([0.5] * num_dim)

edge_utility = np.zeros_like(target_dim_range)

for i, val in enumerate(target_dim_range):
    source_embedding_copy = source_embedding.copy()
    target_embedding_copy = target_embedding.copy()
    source_embedding_copy[target_dim] = val
    target_embedding_copy[target_dim] = val

    edge_embedding = source_embedding_copy * target_embedding_copy

    edge_utility[i] = calc_dimension_utility(target_dim, edge_embedding)

plt.plot(target_dim_range, edge_utility, "--o")
```

```python
edge_utility = np.zeros_like(target_dim_range)

for i, val in enumerate(target_dim_range):
    source_embedding_copy = source_embedding.copy()
    target_embedding_copy = target_embedding.copy()
    source_embedding_copy[target_dim] = val
    target_embedding_copy[target_dim] = 0.5

    edge_embedding = source_embedding_copy * target_embedding_copy

    edge_utility[i] = calc_dimension_utility(target_dim, edge_embedding)

plt.plot(target_dim_range, edge_utility, "--o")
```

```python
edge_utility = np.zeros_like(target_dim_range)

dim_slice = np.concatenate((np.arange(0, target_dim), np.arange(target_dim+1, num_dim)))

similarity_all = np.zeros_like(target_dim_range)
similarity_sub = np.zeros_like(target_dim_range)

for i, val in enumerate(target_dim_range):
    source_embedding_copy = source_embedding.copy()
    target_embedding_copy = target_embedding.copy()
    source_embedding_copy[target_dim] = val
    target_embedding_copy[target_dim] = val

    edge_embedding = source_embedding_copy * target_embedding_copy

    similarity_all[i] = np.mean(edge_embedding)
    similarity_sub[i] = np.mean(edge_embedding[dim_slice])

    edge_utility[i] = calc_dimension_utility(target_dim, edge_embedding)

fig, ax = plt.subplots(1, 1)

ax.plot(target_dim_range, edge_utility, "--o")
ax.plot(target_dim_range, similarity_all, "--o")
ax.plot(target_dim_range, similarity_sub, "--o")
```
